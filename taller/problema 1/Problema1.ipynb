```python
import pandas as pd
```

```python
df=pd.read_csv("./action.csv")
df
```

```python
dictionary = df ["movie_name"]
dictionary
```

```python
import difflib

# Define el vocabulario técnico
vocabulario = dictionary.values
vocabulario = [str(item).lower() for item in vocabulario]

def encontrar_palabra_similar(palabra, vocabulario):
    sugerencia = difflib.get_close_matches(palabra, vocabulario, n=1, cutoff=0.6)

    if sugerencia:
        return sugerencia[0]
    else:
        return "No se encontró una palabra similar en el vocabulario"
while True: 
    # Ejemplo de uso
    palabra_usuario = input("Ingresa la palabra: ")  # La palabra tiene un error ortográfico
    palabra_corregida = encontrar_palabra_similar(palabra_usuario, vocabulario)
    print(f"Palabra corregida: {palabra_corregida}")

```
