```python
import numpy as np
```

```python
def calcular_norma(matriz):
    # Se generan los test cases con vectores canónicos
    test_cases = [[1 if i == j else 0 for j in range(len(matriz))] for i in range(len(matriz))]

    valores_p = [1,2,3,np.inf]
    for p in valores_p:
        dimension = len(matriz)

        # Generar un vector aleatorio con valores en una distribución normal estándar

        for i in range(10000):
            vector_aleatorio = np.random.randn(dimension)

            # Calcular la norma euclidiana del vector
            norma = np.linalg.norm(vector_aleatorio, ord = p)

            # Normalizar el vector dividiendo por su norma
            vector_normalizado = vector_aleatorio / norma
            test_cases.append(vector_normalizado)

        max = 0

        for vec in test_cases:
            producto = np.dot(matriz,vec)
            resultado = np.linalg.norm(producto, ord = p)

            if (max < resultado):
                max = resultado

        print("Norma Orden ",p,": ",max)
```

```python
import numpy as np

n = 10  # Tamaño de la matriz de Hilbert (10x10 en este caso)

# Función para calcular la norma infinito de una matriz
def calcular_norma_infinito(matriz):
    maximo_valor_absoluto = np.max(np.sum(np.abs(matriz), axis=1))
    return maximo_valor_absoluto

# Genera una matriz de Hilbert 10x10
matriz_hilbert = np.zeros((n, n))
for i in range(1, n + 1):
    for j in range(1, n + 1):
        matriz_hilbert[i - 1, j - 1] = 1 / (i + j - 1)

# Calcula la norma infinito de la matriz de Hilbert
norma_infinito = calcular_norma_infinito(matriz_hilbert)

print("Norma infinito de la matriz de Hilbert 10x10:", norma_infinito)
```
